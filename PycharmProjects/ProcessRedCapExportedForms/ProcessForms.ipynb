```python
import numpy as np
import pandas as pd
import os
import pathlib


from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
def filter_exports(fp):

    fp = pathlib.PureWindowsPath(fp).as_posix()
    df = pd.read_csv(fp)

    df = df[df['Survey Timestamp'].notnull()]
    df = df.drop_duplicates(subset='Record ID',keep='first')
    df['Participant ID'] = df['Participant ID'].astype(int).astype(str)

    return df

dde_fp = r"C:\Users\CDT - Admin\Downloads\WST Expectations.csv"
dds_fp = r"C:\Users\CDT - Admin\Downloads\WST Diabetes Distress Scale (DDS).csv"
ddd_fp = r"C:\Users\CDT - Admin\Downloads\WST Demographics.csv"
da_fp = r"C:\Users\CDT - Admin\Downloads\PRODRSOMWST_DATA_LABELS_2021-02-25_1148.csv"
```

```python
dds_df = filter_exports(dds_fp)
dds_df.head(10)

dds_ids_df = pd.DataFrame(dds_df[['Record ID','Participant ID','Participant codeword']])
```

```python
dde_df = filter_exports(dde_fp)
dde_df.head()

ddd_df = filter_exports(ddd_fp)

da_df = filter_exports(da_fp)
```
